```python
"""
creat a class called employee=>
attributes -> company,name,salary,job_designation
hear all the objects of this company will have the same common company
make salary as private veriable

create 3 object => e1,e2,e3 with given attributes

method => 1.methode to calculat the anual salary
          2.method to show entire data
""";
```

```python
class employee:
    company= "mitu"
    def __init__(self,name,salary,job_designation):
        self.name=name
        self.__salary=salary
        self.job_designation=job_designation

    def show1_info(self):
        return self.__salary * 12
        
    def show_info(self):
        print(f" name => {self.name}")
        print(f"salary =>{self.__salary}")
        print(f"company => {self.company}")
        print(f"job_designation => {self.job_designation}")
        print(f"anual => {self.show1_info()}")
```

```python
e1 = employee('sukanya',90000,'moshi')
e2 = employee('samruddhi',90000,'moshi')
e3 = employee('adti',90000,'moshi')

```
