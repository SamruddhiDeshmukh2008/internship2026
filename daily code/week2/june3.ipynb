```python
def vowel_counter(s): 
    vowels = {'a':0 , 'e':0 ,'i':0, 'o':0,'u':0}
    total = 0
    for ch in s.lower():
         if ch in vowels: 
             vowels[ch] += 1 
             total += 1 
    print ("total vowels:",total)
    
    print("vowel counts:",vowels) 
vowel_counter("hello world")
```
