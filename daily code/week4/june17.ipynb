```python
try:
    f= open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("filr dose not exist")
```

```python
try:
    f= open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("filr dose not exist")
except Exception:
    print("something went wrong")
```

```python
try:
    f= open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("filr dose not exist")
except Exception as e:
    print(e)
```

```python
# else -> if try block raise any exception this will run
```

```python
try:
    f= open('a1.txt')
except FileNotFoundError:
    print("filr dose not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
```

```python
# finally
```

```python
try:
    f= open('a1.txt')
    print(var1) 
except FileNotFoundError:
    print("filr dose not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run complete")
```

```python
try:
    f= open('a1.txt')
    if f.name =='a1.txt':
        raise Exception
except FileNotFoundError:
    print("filr dose not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run complete")
```
