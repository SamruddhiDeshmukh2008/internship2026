# Section A — Strings

```python
"""
1. Write a function analyze_sentence(sentence) that takes a sentence and returns a dictionary with
word_count, char_count (excluding spaces), unique_words (case-insensitive), and
most_frequent_word. Test it with:
"The quick brown fox jumps over the lazy dog the fox"
""";
```

```python
def analyze_sentence(sentence):
    words = sentence.lower().split()
    word_count = len(words)
    char_count = len(sentence.replace(" ", ""))
    unique_words = set(words)
    freq = {}
    for w in words:
        freq[w] = freq.get(w, 0) + 1
    most_frequent_word = max(freq, key=freq.get)

    return {
        "word_count": word_count,
        "char_count": char_count,
        "unique_words": unique_words,
        "most_frequent_word": most_frequent_word
    }
print(analyze_sentence("The quick brown fox jumps over the lazy dog the fox"))
```

```python
"""
4. Write a function list_ops(nums) that:
a) Removes duplicates while preserving the original order.
b) Returns a new list with only elements at even indices, squared.
c) Flattens the nested list into a single list using list comprehension.
nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5] nested = [[1, 2, 3], [4, 5], [6, 7, 8,
9]]
""";
```

```python
def list_ops(nums, nested):
    seen = set()
    no_duplicates = []
    for n in nums:
        if n not in seen:
            seen.add(n)
            no_duplicates.append(n)
    even_index_squared = [no_duplicates[i]**2 for i in range(0, len(no_duplicates), 2)]
    flattened = [item for sublist in nested for item in sublist]
    return no_duplicates, even_index_squared, flattened
nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
result = list_ops(nums, nested)
print("No duplicates:", result[0])
print("Even indices squared:", result[1])
print("Flattened list:", result[2])
```

```python
"""
7. Write a function inventory_manager(records) that takes a list of tuples (product, category,
quantity, price) and returns a dictionary grouped by category. Each category should have
total_items, total_value, and a list of product names. Use:
records = [ ("Laptop", "Electronics", 10, 55000), ("Mouse", "Electronics", 50,
499), ("Desk", "Furniture", 8, 12000), ("Chair", "Furniture", 15, 4500),
("Pen", "Stationery", 200, 20), ]
""";
```

```python
def inventory_manager(records):
    inventory = {}
    for product, category, quantity, price in records:
        if category not in inventory:
            inventory[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }
        inventory[category]["total_items"] += quantity
        inventory[category]["total_value"] += quantity * price
        inventory[category]["products"].append(product)
    return inventory

records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20),
]
result = inventory_manager(records)
print(result)
```

```python
"""
8. Implement the following:
a) A function power_all(*args, exp=2) that returns a list of each number raised to exp.
b) A function profile(**kwargs) that prints each key-value pair as Key : Value.
c) Use a lambda inside sorted() to sort the list of dicts below by cgpa in descending order.
students = [{"name": "Arjun", "cgpa": 8.4}, {"name": "Sanya", "cgpa": 9.1},
{"name": "Dev", "cgpa": 7.8}]
""";
```

```python
def power_all(*args, exp=2):
    return [num ** exp for num in args]

def profile(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

students = [
    {"name": "Arjun", "cgpa": 8.4},
    {"name": "Sanya", "cgpa": 9.1},
    {"name": "Dev", "cgpa": 7.8}
]

sorted_students = sorted(students, key=lambda x: x["cgpa"], reverse=True)

print("a) Power All:", power_all(2, 3, 4, exp=3))   # [8, 27, 64]

print("\nb) Profile:")
profile(Name="Riya Sharma", Department="Data Science", Location="Pune")

print("\nc) Students sorted by CGPA (descending):")
for s in sorted_students:
    print(s)
```

```python
"""
9. Create a module stats_utils.py with three functions:
a) mean(data) — arithmetic mean.
b) variance(data) — population variance without using the statistics module.
c) normalize(data) — min-max normalization returning values between 0 and 1.
In a separate main.py, import and use these functions on the dataset below. Also compute the
standard deviation using the math module.
dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]
""";
```

```python
import math
def mean(data):
    return sum(data) / len(data)

def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / len(data)   
def normalize(data):
    min_val, max_val = min(data), max(data)
    return [(x - min_val) / (max_val - min_val) for x in data]

dataset = [12, 45, 7, 23, 56, 89, 34, 67, 11, 90]

print("Mean:", mean(dataset))
print("Variance:", variance(dataset))
print("Normalized:", normalize(dataset))

std_dev = math.sqrt(variance(dataset))
print("Standard Deviation:", std_dev)
```

```python
"""
10. Complete the following:
a) Write a closure make_multiplier(n) that returns a function multiplying its input by n. Create triple
and quintuple using it.
b) Write a decorator @timer that prints the execution time of any function it wraps. Apply it to
slow_sum(n) which computes the sum of squares from 1 to n.
c) Write a decorator @validate_positive that raises a ValueError if any argument is zero or negative.
Apply it to safe_sqrt(x).
""";
```

```python
import time
import math

def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier

triple = make_multiplier(3)
quintuple = make_multiplier(5)

print("Triple of 10:", triple(10))       
print("Quintuple of 4:", quintuple(4)) 

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"Execution time: {end - start:.5f} seconds")
        return result
    return wrapper

def slow_sum(n):
    return sum(i**2 for i in range(1, n+1))

print("Slow sum result:", slow_sum(10000))

def validate_positive(func):
    def wrapper(*args, **kwargs):
        for arg in args:
            if arg <= 0:
                raise ValueError("All arguments must be positive")
        return func(*args, **kwargs)
    return wrapper

def safe_sqrt(x):
    return math.sqrt(x)

print("Square root of 16:", safe_sqrt(16))
```
