```python
"""
1. Create a class BankAccount with the following:
a) Attributes: account_holder, account_number, and balance.
b) Methods: deposit(amount), withdraw(amount), and get_balance().
c) withdraw() should print a message and do nothing if the amount exceeds the current balance.
d) A __str__ method that returns a readable summary of the account.
Create two BankAccount objects and perform a few deposits and withdrawals
""";
```

```python
class BankAccount:
    
    def __init__(self, account_holder, account_number, balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"₹{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance! Withdrawal not possible.")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully.")

    def get_balance(self):
        return self.balance

    def __str__(self):
        return (f"Account Holder: {self.account_holder}\n"
                f"Account Number: {self.account_number}\n"
                f"Balance: ₹{self.balance}")
```

```python
acc1 = BankAccount("samruddhi", "ACC001", 5000)
acc2 = BankAccount("aditi", "ACC002", 10000)

acc1.deposit(2000)
acc1.withdraw(8000)
acc2.withdraw(12000)
print(acc1)
print(acc2)
```

```python
"""
2. Create a class Student with attributes name, roll_no, and marks (a list of 5 subject marks). Add
methods:
a) calculate_percentage() — returns the average of all marks.
b) get_grade() — returns 'A' for >= 80, 'B' for >= 60, 'C' for >= 40, and 'Fail' otherwise.
c) display() — prints all details including percentage and grade.
Create a list of 3 Student objects and call display() on each.
""";
```

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks 

    def calculate_percentage(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        percentage = self.calculate_percentage()
        if percentage >= 80:
            return "A"
        elif percentage >= 60:
            return "B"
        elif percentage >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        percentage = self.calculate_percentage()
        grade = self.get_grade()
        print(f"Name: {self.name}, Roll No: {self.roll_no}, Marks: {self.marks}, "
              f"Percentage: {percentage:.2f}, Grade: {grade}")


# Example usage
students = [
    Student("samruddhi", "S001", [85, 90, 78, 88, 92]),
    Student("sukanya", "S002", [65, 70, 60, 58, 72]),
    Student("adti", "S003", [35, 40, 38, 42, 45])
]

for s in students:
    s.display()
```

```python
"""
3. Create a class Employee with private attributes __name, __emp_id, and __salary. Add:
a) Getter and setter methods for all three attributes.
b) The salary setter should raise a ValueError if the value is below 10000.
c) A method give_raise(percent) that increases salary by the given percentage.
Demonstrate that the private attributes cannot be accessed directly, and that all access goes
through the methods.
""";
```

```python
class Employee:
    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.__salary = salary

    # Getters
    def get_name(self):
        return self.__name

    def get_emp_id(self):
        return self.__emp_id

    def get_salary(self):
        return self.__salary

    # Setters
    def set_name(self, name):
        self.__name = name

    def set_emp_id(self, emp_id):
        self.__emp_id = emp_id

    def set_salary(self, salary):
        if salary < 10000:
            raise ValueError("Salary must be at least 10000")
        self.__salary = salary

    # Raise method
    def give_raise(self, percent):
        self.__salary += self.__salary * (percent / 100)

emp = Employee("samruddhi", "E042", 72000)
print(emp.get_name(), emp.get_emp_id(), emp.get_salary())

emp.give_raise(10)
print("aditi:", emp.get_salary())
```

```python
"""
4. Build a class SecureVault that stores a private __pin and a private __data dictionary. It should
have:
a) A method add_entry(pin, key, value) that adds to __data only if the pin matches.
b) A method get_entry(pin, key) that returns the value only if the pin matches.
c) Any wrong pin attempt should print 'Access Denied' and return nothing.
""";
```

```python
class SecureVault:
    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):
        if pin == self.__pin:
            self.__data[key] = value
            print("Entry added successfully.")
        else:
            print("Access Denied")

    def get_entry(self, pin, key):
        if pin == self.__pin:
            return self.__data.get(key, None)
        else:
            print("Access Denied")
            return None

vault = SecureVault("1234")
vault.add_entry("1234", "email", "riya@example.com")
vault.add_entry("9999", "phone", "9876543210")
print(vault.get_entry("1234", "email"))        
print(vault.get_entry("0000", "email"))        
```

```python
"""
5. Create a base class Shape with a method area() that returns 0 and a method describe() that
prints 'I am a shape'. Create three subclasses:
a) Circle(radius) — overrides area() using pi * r^2.
b) Rectangle(length, width) — overrides area().
c) Triangle(base, height) — overrides area().
Each subclass should also override describe() to print something specific about itself. Create one
object of each and call both methods
""";
```

```python
import math

class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am a shape")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius**2

    def describe(self):
        print("I am a circle with radius", self.radius)

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def describe(self):
        print("I am a rectangle with length", self.length, "and width", self.width)

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

    def describe(self):
        print("I am a triangle with base", self.base, "and height", self.height)


shapes = [Circle(5), Rectangle(4, 6), Triangle(3, 7)]
for s in shapes:
    s.describe()
    print("Area:", s.area())
```

```python
"""
6. Create a class Vehicle with attributes make, model, and year and a method start_engine() that
prints 'Engine started'. Create two subclasses:
a) Car(make, model, year, num_doors) — overrides start_engine() to include the car's name in
the message.
b) ElectricCar(make, model, year, battery_capacity) — overrides start_engine() to print 'Silent
start — battery at X kWh'.
Use super() in both subclasses to call the parent __init__. Create objects and test all methods
""";
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def start_engine(self):
        print(f"Engine started for {self.make} {self.model} ({self.num_doors}-door)")

class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print(f"Silent start — battery at {self.battery_capacity} kWh")


car = Car("Toyota", "Corolla", 2022, 4)
ev = ElectricCar("Tesla", "Model 3", 2023, 75)

car.start_engine()
ev.start_engine()
```

```python
"""7. Create a base class Animal with a method sound() that prints 'Some sound'. Create subclasses
Dog, Cat, and Cow, each overriding sound() with their respective sounds. Write a function
make_all_sounds(animals) that takes a list of Animal objects and calls sound() on each.
Demonstrate that the correct method is called regardless of the object type.
""";
```

```python
class Animal:

    def sound(self):
        print("Some sound")


class Dog(Animal):

    def sound(self):
        print("Bark")


class Cat(Animal):

    def sound(self):
        print("Meow")


class Cow(Animal):

    def sound(self):
        print("Moo")


def make_all_sounds(animals):
    for animal in animals:
        animal.sound()
```

```python
dog = Dog()
cat = Cat()
cow = Cow()
animals = [dog, cat, cow]
make_all_sounds(animals)
```

```python
"""
8. Create a class Calculator with overloaded-style methods using default arguments:
a) add(a, b, c=0) — adds two or three numbers.
b) multiply(a, b, c=1) — multiplies two or three numbers.
c) Also implement __add__ on a class Vector(x, y) so that two Vector objects can be added with
the + operator. Print the resulting vector
""";
```

```python
class Calculator:

    def add(self, a, b, c=0):
        return a + b + c

    def multiply(self, a, b, c=1):
        return a * b * c


class Vector:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"
calc = Calculator()

print("Addition of 2 numbers:", calc.add(10, 20))
print("Addition of 3 numbers:", calc.add(10, 20, 30))

print("Multiplication of 2 numbers:", calc.multiply(5, 4))
print("Multiplication of 3 numbers:", calc.multiply(5, 4, 2))

v1 = Vector(2, 3)
print(v1)
```

```python
"""
10. Build a menu-driven Student Record Manager using OOP and file handling. The app should
have a class StudentManager with the following menu options:
1. Add Student — takes name, roll number, and marks for 3 subjects.
2. View All — displays all stored students with their average marks.
3. Search Student — finds a student by roll number.
4. Delete Student — removes a student by roll number.
5. Save to File — writes all records to students_data.txt.
6. Load from File — reads records back from students_data.txt on startup.
7. Exit
Requirements: use a class with a list as internal storage, handle invalid menu choices gracefully,
and use a while loop to keep the menu running until the user chooses to exit.
""";
```

```python
import os

class StudentManager:
    def __init__(self):
        self.students = []
        self.filename = "students_data.txt"
        self.load_from_file()

    def add_student(self, name, roll_no, marks):
        avg = sum(marks) / len(marks)
        self.students.append({"name": name, "roll_no": roll_no, "marks": marks, "average": avg})
        print("Student added successfully!")

    def view_all(self):
        if not self.students:
            print("No records found.")
        else:
            for s in self.students:
                print(f"Roll No: {s['roll_no']}, Name: {s['name']}, Marks: {s['marks']}, Average: {s['average']:.2f}")

    def search_student(self, roll_no):
        for s in self.students:
            if s["roll_no"] == roll_no:
                print(f"Found → Roll No: {s['roll_no']}, Name: {s['name']}, Average: {s['average']:.2f}")
                return
        print("Student not found.")

    def delete_student(self, roll_no):
        for s in self.students:
            if s["roll_no"] == roll_no:
                self.students.remove(s)
                print("Student deleted successfully!")
                return
        print("Student not found.")

    def save_to_file(self):
        with open(self.filename, "w") as f:
            for s in self.students:
                line = f"{s['name']},{s['roll_no']},{','.join(map(str, s['marks']))}\n"
                f.write(line)
        print("Records saved to file.")

    def load_from_file(self):
        if os.path.exists(self.filename):
            with open(self.filename, "r") as f:
                for line in f:
                    parts = line.strip().split(",")
                    if len(parts) >= 5:
                        name, roll_no = parts[0], parts[1]
                        marks = list(map(int, parts[2:]))
                        avg = sum(marks) / len(marks)
                        self.students.append({"name": name, "roll_no": roll_no, "marks": marks, "average": avg})
            print("Records loaded from file.")


# ---------------- MENU LOOP ----------------
def main():
    manager = StudentManager()
    while True:
        print("\n--- Student Record Manager ---")
        print("1. Add Student")
        print("2. View All")
        print("3. Search Student")
        print("4. Delete Student")
        print("5. Save to File")
        print("6. Load from File")
        print("7. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            name = input("Enter name: ")
            roll_no = input("Enter roll number: ")
            marks = [int(input(f"Enter marks for subject {i+1}: ")) for i in range(3)]
            manager.add_student(name, roll_no, marks)

        elif choice == "2":
            manager.view_all()

        elif choice == "3":
            roll_no = input("Enter roll number to search: ")
            manager.search_student(roll_no)

        elif choice == "4":
            roll_no = input("Enter roll number to delete: ")
            manager.delete_student(roll_no)

        elif choice == "5":
            manager.save_to_file()

        elif choice == "6":
            manager.load_from_file()

        elif choice == "7":
            print("Exiting... Goodbye!")
            break

        else:
            print("Invalid choice. Please try again.")


# Run the app
# main()
```

```python
import os

class StudentManager:

    def __init__(self):
        self.students = []

    def add_student(self):
        name = input("Enter Name: ")
        roll_no = input("Enter Roll No: ")

        marks = []
        for i in range(3):
            mark = float(input(f"Enter Marks of Subject {i+1}: "))
            marks.append(mark)

        student = {
            "name": name,
            "roll_no": roll_no,
            "marks": marks
        }

        self.students.append(student)
        print("Student Added Successfully!")

    def view_all(self):
        if len(self.students) == 0:
            print("No Records Found!")
            return

        for student in self.students:
            avg = sum(student["marks"]) / 3

            print("\nName:", student["name"])
            print("Roll No:", student["roll_no"])
            print("Marks:", student["marks"])
            print("Average:", avg)

    def search_student(self):
        roll = input("Enter Roll No to Search: ")

        for student in self.students:
            if student["roll_no"] == roll:
                avg = sum(student["marks"]) / 3

                print("\nStudent Found")
                print("Name:", student["name"])
                print("Roll No:", student["roll_no"])
                print("Marks:", student["marks"])
                print("Average:", avg)
                return

        print("Student Not Found!")

    def delete_student(self):
        roll = input("Enter Roll No to Delete: ")

        for student in self.students:
            if student["roll_no"] == roll:
                self.students.remove(student)
                print("Student Deleted Successfully!")
                return

        print("Student Not Found!")

    def save_to_file(self):
        file = open("students_data.txt", "w")

        for student in self.students:
            line = (
                student["name"] + "," +
                student["roll_no"] + "," +
                str(student["marks"][0]) + "," +
                str(student["marks"][1]) + "," +
                str(student["marks"][2]) + "\n"
            )
            file.write(line)

        file.close()
        print("Records Saved Successfully!")

    def load_from_file(self):
        if os.path.exists("students_data.txt"):

            file = open("students_data.txt", "r")

            for line in file:
                data = line.strip().split(",")

                student = {
                    "name": data[0],
                    "roll_no": data[1],
                    "marks": [
                        float(data[2]),
                        float(data[3]),
                        float(data[4])
                    ]
                }

                self.students.append(student)

            file.close()

# Main Program
manager = StudentManager()

manager.load_from_file()

while True:

    print("\n===== Student Record Manager =====")
    print("1. Add Student")
    print("2. View All")
    print("3. Search Student")
    print("4. Delete Student")
    print("5. Save to File")
    print("6. Load from File")
    print("7. Exit")

    choice = input("Enter Choice: ")

    if choice == "1":
        manager.add_student()

    elif choice == "2":
        manager.view_all()

    elif choice == "3":
        manager.search_student()

    elif choice == "4":
        manager.delete_student()

    elif choice == "5":
        manager.save_to_file()

    elif choice == "6":
        manager.load_from_file()
        print("Records Loaded Successfully!")

    elif choice == "7":
        manager.save_to_file()
        print("Exiting Program...")
        break

    else:
        print("Invalid Choice! Please Enter 1 to 7.")
```
