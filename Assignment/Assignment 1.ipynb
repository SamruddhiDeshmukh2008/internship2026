```python
#2. Accept a temperature in Celsius from the user, convert it to Fahrenheit and Kelvin, and
#print all three values with labels
```

```python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = (celsius * 9/5) + 32
kelvin = celsius + 273.15

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
print("Kelvin:", kelvin)
```

```python
# 8. Take a positive integer n as input and print all prime numbers from 2 to n.
```

```python
n = int(input("Enter a positive integer: "))

print("Prime numbers up to", n, ":")
for num in range(2, n+1):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        print(num, end=" ")
```

```python
# 9. Keep accepting numbers from the user until they enter 0. Then print the total count ofnumbers entered,
# their sum, and their average. Do not include 0 in the calculations.
```

```python
count = 0
total = 0

while True:
    num = int(input("Enter a number (0 to stop): "))
    if num == 0:
        break
    count += 1
    total += num

if count > 0:
    average = total / count
    print("Count:", count)
    print("Sum:", total)
    print("Average:", average)
else:
    print("No numbers entered.")
```

```python
# 10. Take n as input and print a triangle pattern of n rows made of asterisks, where row i has (2i - 1) stars centered appropriately
```

```python
n = int(input("Enter number of rows: "))

for i in range(1, n+1):
    stars = "*" * (2*i - 1)
    print(stars.center(2*n - 1))
```
