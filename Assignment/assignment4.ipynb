# Section A — Exception Handling


```python
"""
1. Write a function safe_divide(a, b) that divides two numbers and handles ZeroDivisionError and
TypeError separately, printing an appropriate message for each. Use a finally block to print
'Operation attempted.' regardless of the outcome. Test it with valid numbers, division by zero, and a
string input.
"""
```

```python
def safe_divide(a, b):
    try:
        result = a / b
        print(f"Result: {result}")
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
    except TypeError:
        print("Error: Both inputs must be numbers.")
    finally:
        print("Operation attempted.")

# Test cases
safe_divide(10, 2)  # valid division
safe_divide(10, 0)  # division by zero
safe_divide(10, "a")  # invalid type
```

```python
"""
2. Write a function validate_age(age) that raises a custom exception InvalidAgeError (a class you
define, inheriting from Exception) if age is negative or greater than 120. Catch this exception in a
calling block and print a user-friendly message. Test it with at least 3 different values.
""";
```

```python
class InvalidAgeError(Exception):
    pass

def validate_age(age):
    if age < 0 or age > 120:
        raise InvalidAgeError(f"Invalid age: {age}")
    print(f"Age {age} is valid.")

for test_age in [25, -5, 130]:
    try:
        validate_age(test_age)
    except InvalidAgeError as e:
        print(f"Caught an error: {e}")
```

```python
"""
3. Write a program that repeatedly asks the user (use a loop with at least 3 simulated inputs in a list,
e.g. user_inputs = ['10', 'abc', '0', '5']) to enter a number and divide 100 by it. Handle the following in
a single try block using multiple except clauses:
a) ValueError if the input is not a number.
b) ZeroDivisionError if the input is zero.
c) A generic Exception as a catch-all for anything else.
Print the result if successful, otherwise print which error occurred
""";
```

```python
user_inputs = ['10', 'abc', '0', '5']

for inp in user_inputs:
    try:
        num = int(inp)
        result = 100 / num
        print(f"100/{num}:{result}")
    except ValueError:
        print(f"Error :'{inp}' is not a valid number.")
    except ZeroDivisionError:
        print("Error: cannot divide by zero.")
    except Exception as e:
        print(f"Unexpected error: {e}")
```

```python
"""
4. Create a 1D NumPy array of integers from 1 to 20. Perform the following:
a) Reshape it into a 4x5 2D array.
b) Print the sum of each row and each column.
c) Extract all elements greater than 10 using boolean indexing.
d) Replace all even numbers in the array with -1.
""";
```

```python
import numpy as np
arr = np.arange(1, 21)

arr_2d = arr.reshape(4 ,5)
print("4*5 Array:\n",arr_2d)

row_sum = arr_2d.sum(axis=1)
col_sum = arr_2d.sum(axis=0)
print("\n sum of row:", row_sum)
print("\n sum of column:", col_sum)

greater_than_10 = arr[arr > 10]
print("\n greater number than 10:", greater_than_10)

arr_replaced = np.where(arr%2 == 0, -1, arr)
print("/n array witn even number replaced:\n", arr_replaced.reshape(4, 5))
```

```python
"""
6. Generate a NumPy array of 15 random integers between 1 and 100 using np.random.randint()
(set a seed of 42 for reproducibility). Then:
a) Find the mean, median, standard deviation, min, and max.
b) Sort the array in descending order.
c) Find the indices of the 3 largest values using np.argsort().
d) Normalize the array so all values lie between 0 and 1
""";
```

```python
import numpy as np

np.random.seed(42)
arr = np.random.randint(1,101, size=15)

mean = np.mean(arr)
median = np.median(arr)
standard_deviation = np.std(arr)
min_v = np.min(arr)
max_v = np.max(arr)

sorted_desc = np.sort(arr)

indices_top3 = np.argsort(arr)[-3:][::-1]
Normalize = (arr - min_v) / (max_v - min_v)

print("\n Array", arr)
print("\n median", median)
print("\n mean", mean)
print("\n standard_deviation", standard_deviation)
print("\n min_v", min_v)
print("\n max_v", max_v)
print("\n sorted_desc", sorted_desc)
print("\n indices_top3", indices_top3)
print("\n Normalize", Normalize)
```

# Section C — Pandas Basics

```python
"""
7. Create a Pandas DataFrame from the dictionary below and perform the following:
data = { "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"], "Age": [23, 21,
25, 22, 24], "Department": ["AI", "Data Science", "AI", "Web Dev", "Data
Science"], "Score": [88, 92, 79, 95, 67], }
a) Display the first 3 rows and the data types of each column.
b) Add a new column Grade based on Score (>=90 'A', >=75 'B', else 'C').
c) Sort the DataFrame by Score in descending order
""";
```

```python

import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67],
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

print("\nFirst 3 rows:\n", df.head(3))
print("\nData types:\n", df.dtypes)

df["Grade"] = df["Score"].apply(lambda s: "A" if s >= 90 else ("B" if s >= 75 else "C"))
print("\nDataFrame with Grade column:\n", df)

df_sorted = df.sort_values(by="Score", ascending=False)
print("\nDataFrame sorted by Score (descending):\n", df_sorted)

df_ai = df[df["Department"] == "AI"]
print("\nRows where Department is 'AI':\n", df_ai)
```

```python
"""
8. Using the same DataFrame from Q7:
a) Group by Department and find the average Score for each.
b) Find the row with the highest Score using idxmax().
c) Rename the column Score to Marks.
d) Drop the Age column and display the resulting DataFrame.
""";
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67],
}

df = pd.DataFrame(data)

avg_scores = df.groupby("Department")["Score"].mean()
print("Average Score by Department:\n", avg_scores)

highest_row = df.loc[df["Score"].idxmax()]
print("\nRow with highest Score:\n", highest_row)

df.rename(columns={"Score": "Marks"}, inplace=True)
print("\nDataFrame with 'Score' renamed to 'Marks':\n", df)

df_dropped = df.drop(columns=["Age"])
print("\nDataFrame after dropping 'Age':\n", df_dropped)
```

```python
"""
9. Create a DataFrame with some missing values as shown below and handle them:
import numpy as np data = { "Product": ["Pen", "Notebook", "Eraser",
"Sharpener", "Ruler"], "Price": [10, np.nan, 5, 8, np.nan], "Stock": [100, 50,
np.nan, 30, 20], }
a) Check the number of missing values in each column using isnull().sum().
b) Fill missing Price values with the column mean.
c) Drop rows where Stock is still missing.
""";
```

```python
import pandas as pd
import numpy as np

data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20],
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

print("\nMissing values per column:\n", df.isnull().sum())

mean_price = df["Price"].mean()
df["Price"].fillna(mean_price, inplace=True)
print("\nDataFrame after filling missing Price with mean:\n", df)

df_cleaned = df.dropna(subset=["Stock"])
print("\nDataFrame after dropping rows with missing Stock:\n", df_cleaned)

df_cleaned.to_csv("cleaned_products.csv", index=False)
print("\n✅ Cleaned DataFrame saved to 'cleaned_products.csv'")
```
